# Create CSV with Neighboring Countries

```python
import pandas as pd
import geopandas as gpd
```

```python
borderbuffer = gpd.read_file('borderbuffer_streets_5km_eunuts.shp')
```

```python
borderbuffer.head()
```

```python
countries = list(borderbuffer.country1.unique())
countries.extend(list(borderbuffer.country2.unique()))
countries = set(countries)
countries
```

```python
dict_ = {}
for c in countries:
    tab = borderbuffer[(borderbuffer.country1 == c) | (borderbuffer.country2 == c)].copy()
    adj = len(tab.border.unique())
    #dict_.update({c:adj})
    dict_.update({c: {'country':c, 'neighbors': adj}})
df = pd.DataFrame.from_dict(dict_, orient = 'index')

df.to_csv('neighboring_countries.csv')
df
```
